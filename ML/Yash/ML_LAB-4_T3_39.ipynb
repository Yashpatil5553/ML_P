### Preprocessing steps

```python
# step 1 : Separate input and output variable
X= dataset.iloc[:,:-1].values
Y= dataset.iloc[:,-1].values
```

```python
print(X)
print(Y)
```

```python
# Step 4 : splitting the data into training and testing

from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.30,random_state=39)
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
# Step 5: Feature Scaling
from sklearn.preprocessing import StandardScaler
sc= StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

```python
print(X_test)
```

### Training the decision tree classification model on the training set

```python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(criterion = 'entropy', random_state = 39)
classifier.fit(X_train, Y_train)
```

### Predicting the test result

```python
ypred = classifier.predict(X_test)
```
